```python
#This is a program to scrape lyrics using Genius Song API.It calls the API by inputting Song name and Artist name.
```

```python
import pandas as pd

import time

from lyricsgenius import Genius
from requests.exceptions import Timeout
```

```python
genius = Genius('Q8l531GAQ_fD0aXPT0Xf-GlKxLAbeZhjp_k1FCIKZ3rwItea-uVxmPZQrhLT_oqm')
```

```python
df = pd.read_csv('dataset-of-90s-Song-4001-5521.csv')
df.head()
```

```python
tracks = df['track'].tolist()
artists = df['artist'].tolist()
```

```python
a_zip = zip(tracks, artists)
trackArtist = list(a_zip)
trackArtist
```

```python
trackArtistLen = len(trackArtist)
print(trackArtistLen)
lyricList = []
```

```python
def get_lyrics():
    start_time = time.time()

    genius.timeout = 120 
    genius.sleep = 5
    songCounter=1
    for pair in trackArtist:# for each pair of track and artist

        retries = 0 # retry counter
        while retries < 3: # while retries are less than 3
            try:
                song = genius.search_song(pair[0], pair[1])
                try:
                    lyricList.append(song.lyrics)
                except:
                    lyricList.append('No lyrics')

            except Timeout as e: # if timeout
                retries += 1 # add to counter and try again
                continue
            
            break
        #Printing a statement at each interval of 100 multiples for informational purpose
        if(songCounter%100 == 0):
            print('####################################') 
            print('song:', songCounter) 
            print('####################################') 
        songCounter += 1
    seconds = time.time() - start_time
    print('Time Taken:', time.strftime("%H:%M:%S",time.gmtime(seconds)))

```

```python
get_lyrics()
```

```python
# add the lyrics list to the dataframe
df['lyrics'] = lyricList
# save the dataframe to CSV
df.to_csv('dataset-of-00s-song-1-2000-with-lyrics-empty.csv', sep=',')
```
